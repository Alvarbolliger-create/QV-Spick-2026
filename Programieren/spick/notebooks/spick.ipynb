### <a id="Plattenkondenstor"></a> Plattenkondensator

```python
import ipywidgets as widgets
from IPython.display import display

# ---- Module importieren ----
from Formeln.Cap import platte_berechnen
from Utils.Parsing import parse_value as p  # SI-Präfixe unterstützen

# ----------------------------
# Berechnungsfunktion
# ----------------------------
def kondensator_rechner(_=None):
    out_C.clear_output()
    with out_C:
        try:
            # Werte parsen, leere Felder → None
            C = p(C_input_C.value) if C_input_C.value.strip() else None
            A = p(A_input_C.value) if A_input_C.value.strip() else None
            d = p(d_input_C.value) if d_input_C.value.strip() else None
            eps_r = p(eps_r_input_C.value) if eps_r_input_C.value.strip() else None

            values = [C, A, d, eps_r]
            none_count = sum(x is None for x in values)

            if none_count == 0:
                print("❌ Bitte genau ein Feld leer lassen, um es zu berechnen!")
                return
            elif none_count > 1:
                print("❌ Bitte nicht mehr als ein Feld leer lassen!")
                return

            # Jetzt genau ein Feld None → Berechnung möglich
            ergebnis = platte_berechnen(C=C, A=A, d=d, eps_r=eps_r)

        except Exception as e:
            print("❌ Ungültige Eingabe:", e)
            return

        print("Berechnete Werte:")
        print(f"C     = {ergebnis['C']:.6e} F")
        print(f"A     = {ergebnis['A']:.6e} m²")
        print(f"d     = {ergebnis['d']:.6e} m")
        print(f"ε_r   = {ergebnis['eps_r']:.6g}")

# ----------------------------
# Theorie-Box
# ----------------------------
theorie_html = widgets.HTML("""
<p>Die Kapazität eines Plattenkondensators wird berechnet mit:</p>
<p style="text-align:center; font-weight:bold;">C = ε_r · ε0 · A / d</p>
<ul>
<li>[C] = Kapazität in Farad (F)</li>
<li>[A] = Fläche der Platten (m²)</li>
<li>[d] = Abstand der Platten (m)</li>
<li>ε0 ≈ 8.854·10⁻¹² F/m</li>
<li>ε_r = Dielektrizitätszahl des Materials</li>
</ul>
""")

theorie_box = widgets.VBox([theorie_html])
theorie_box.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='300px',
    display='flex',
    align_items='flex-start'
)

# ----------------------------
# Eingabefelder (Rechner)
# ----------------------------
C_input_C = widgets.Text(value="", description="C [F]:")
A_input_C = widgets.Text(value="", description="A [m²]:")
d_input_C = widgets.Text(value="", description="d [m]:")
eps_r_input_C = widgets.Text(value="", description="ε_r:")

out_C = widgets.Output(layout=widgets.Layout(min_height='130px'))

# ----------------------------
# Widgets beobachten
# ----------------------------
for w in [C_input_C, A_input_C, d_input_C, eps_r_input_C]:
    w.observe(kondensator_rechner, names='value')

# ----------------------------
# Rechner-Box Layout
# ----------------------------
rechner_box = widgets.VBox([C_input_C, A_input_C, d_input_C, eps_r_input_C, out_C])
rechner_box.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='300px',
    display='flex',
    flex_flow='column',
    align_items='flex-start'
)

# ----------------------------
# Gesamt-Layout nebeneinander
# ----------------------------
layout = widgets.HBox([theorie_box, rechner_box],
                      layout=widgets.Layout(
                          justify_content='center',
                          align_items='flex-start',
                          width='100%'
                      ))

display(layout)

# ----------------------------
# Initialer Aufruf für leeren Rechner
# ----------------------------
kondensator_rechner()
```

### <a id="Kondastor_Lade_Vorgang"></a> Kondensator Ladevorgang

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# ---- Module importieren ----
from Formeln.ResCap import rc_tau, u_lade, u_entlade, q_lade
from Utils.Parsing import parse_value as p  # Alias

# ----------------------------
# Interaktive Berechnung
# ----------------------------
def rc_rechner(_=None):
    out_RC.clear_output()
    with out_RC:
        try:
            R = p(R_input_RC.value)
            C = p(C_input_RC.value)
            U0 = p(U0_input_RC.value)
        except:
            print("❌ Ungültige Eingabe! Zahlen oder SI-Präfixe verwenden.")
            return

        if None in (R, C, U0):
            # Wenn keine Eingabe, verwenden wir Standardwerte für initialen Plot
            R = R or 1000
            C = C or 1e-6
            U0 = U0 or 5

        tau = rc_tau(R, C)
        t_max = 5*tau
        t_vals = np.linspace(0, t_max, 300)
        U_l = u_lade(t_vals, R, C, U0)
        U_e = u_entlade(t_vals, R, C, U0)

        try:
            if mode_input_RC.value == "Zeit → Spannung":
                t_user = p(t_input_RC.value) or 0
                t_user = max(0, min(t_user, t_max))
                U_user = u_lade(t_user, R, C, U0)
                label_str = f"t={t_user:.6g}s → U={U_user:.3f}V"
            else:
                U_user = p(U_input_RC.value) or 0
                U_user = max(0, min(U_user, U0))
                t_user = -tau * np.log(1 - U_user/U0) if U_user < U0 else t_max
                label_str = f"U={U_user:.3f}V → t={t_user:.6g}s"
        except Exception as e:
            print("❌ Fehler bei Berechnung:", e)
            return

        Q_user = q_lade(C, U_user)

        plt.figure(figsize=(6,4))
        plt.plot(t_vals, U_l, label="Lade-Kurve")
        plt.plot(t_vals, U_e, label="Entlade-Kurve")
        plt.axvline(t_user, color='gray', linestyle='--', label=label_str)
        plt.scatter(t_user, U_user, color='red', zorder=5)
        plt.title(f"RC-Kreis: R={R:.3g} Ω, C={C:.3g} F, U0={U0:.3g} V (τ={tau:.3g} s)")
        plt.xlabel("Zeit [s]")
        plt.ylabel("Spannung [V]")
        plt.xlim(0, t_max)
        plt.ylim(0, 1.1*U0)
        plt.grid(True)
        plt.legend()
        plt.show()
        print(f"Zeitkonstante τ = {tau:.6g} s")
        print(label_str)
        print(f"Gespeicherte Ladung Q = {Q_user:.6g} C")

# ----------------------------
# Output für Plot
# ----------------------------
out_RC = widgets.Output(layout=widgets.Layout(min_height='300px'))  # fester Platz für Plot

# ----------------------------
# Theorie-Box
# ----------------------------
theorie_html = widgets.HTML("""
<p>Die Zeitkonstante τ eines RC-Kreises wird berechnet durch:</p>
<p style="text-align:center; font-weight:bold;">τ = R &middot; C</p>

<ul>
<li>R: Widerstand [Ω]</li>
<li>C: Kapazität [F]</li>
<li>τ: Zeitkonstante [s]</li>
</ul>

<p>Ladespannung am Kondensator:</p>
<p style="text-align:center;">U(t) = U0 &middot; (1 - e<sup>-t/τ</sup>)</p>

<p>Entladungsspannung:</p>
<p style="text-align:center;">U(t) = U0 &middot; e<sup>-t/τ</sup></p>

<hr>

<h4>Ladung & Strom</h4>

<p>Die gespeicherte Ladung Q auf einem Kondensator:</p>
<p style="text-align:center; font-weight:bold;">Q = C &middot; U</p>

<p>Zusammenhang von Strom I und Ladung Q:</p>
<p style="text-align:center; font-weight:bold;">Q = I &middot; t  &nbsp;&nbsp;→&nbsp;&nbsp; I = dQ/dt</p>
""")


theorie_box = widgets.VBox([theorie_html])
theorie_box.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='700px',  # gleiche Höhe wie Rechner-Box
    display='flex',
    align_items='flex-start'
)

# ----------------------------
# Widgets für Rechner
# ----------------------------
R_input_RC = widgets.Text(value="1k", description="R [Ω]:")
C_input_RC = widgets.Text(value="1u", description="C [F]:")
U0_input_RC = widgets.Text(value="5", description="U0 [V]:")

mode_input_RC = widgets.Dropdown(
    options=["Zeit → Spannung", "Spannung → Zeit"],
    value="Zeit → Spannung",
    description="Modus:"
)

t_input_RC = widgets.Text(value="0.0", description="t [s]:")
U_input_RC = widgets.Text(value="2.5", description="U [V]:")

# Output für Plot
out_RC = widgets.Output(layout=widgets.Layout(min_height='300px'))

# ----------------------------
# Input-Layout: 2 Spalten à 3 Widgets
# ----------------------------
left_column = widgets.VBox([R_input_RC, C_input_RC, U0_input_RC])
right_column = widgets.VBox([mode_input_RC, t_input_RC, U_input_RC])

input_columns = widgets.HBox([left_column, right_column],
                             layout=widgets.Layout(
                                 width='100%',
                                 justify_content='space-between'
                             ))

# Gesamte Rechner-Box
rechner_box = widgets.VBox([input_columns, out_RC])
rechner_box.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='700px',
    display='flex',
    flex_flow='column',
    align_items='flex-start'
)

# ----------------------------
# Trigger auf Änderung
# ----------------------------
for w in [R_input_RC, C_input_RC, U0_input_RC, t_input_RC, U_input_RC, mode_input_RC]:
    w.observe(rc_rechner, names='value')

# ----------------------------
# Initialer Plot beim Laden
# ----------------------------
rc_rechner()

# ----------------------------
# Gesamt-Layout: beide Boxen nebeneinander, zentriert
# ----------------------------
layout = widgets.HBox([theorie_box, rechner_box],
                      layout=widgets.Layout(
                          justify_content='center',
                          align_items='flex-start',
                          width='100%'
                      ))

display(layout)
```

### <a id="Zeitkonstante"></a> Zeitkonstante

```python
# ----------------------------
# Import
# ----------------------------
import ipywidgets as widgets
from IPython.display import display
from Utils.Parsing import parse_value as p
from Formeln.ResCap import rc_tau

# ----------------------------
# Widgets für Rechner
# ----------------------------
R_input_tau = widgets.Text(description="R [Ω]:", value="10k")
C_input_tau = widgets.Text(description="C [F]:", value="100n")
out_tau = widgets.Output(layout=widgets.Layout(min_height='150px'))

# ----------------------------
# Observer-Funktion
# ----------------------------
def update_tau(_=None):
    out_tau.clear_output()
    with out_tau:
        try:
            R = p(R_input_tau.value) if R_input_tau.value.strip() else None
            C = p(C_input_tau.value) if C_input_tau.value.strip() else None
            tau = rc_tau(R, C)
            print(f"τ = {tau:.4e} s")
        except Exception as e:
            print("❌ Ungültige Eingabe:", e)

# Observer hinzufügen
for w in [R_input_tau, C_input_tau]:
    w.observe(update_tau, names='value')

# ----------------------------
# Rechner-Box
# ----------------------------
rechner_box_tau = widgets.VBox([R_input_tau, C_input_tau, out_tau])
rechner_box_tau.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='300px',
    display='flex',
    flex_flow='column',
    align_items='flex-start'
)

# ----------------------------
# Theorie-Box
# ----------------------------
theorie_html_tau = widgets.HTML("""
<p>Die Zeitkonstante τ eines RC-Kreises wird berechnet mit:</p>
<p style="text-align:center; font-weight:bold;">τ = R · C</p>
<ul>
<li>R: Widerstand [Ω]</li>
<li>C: Kapazität [F]</li>
<li>τ: Zeitkonstante [s]</li>
</ul>
<p>- Nach τ Sekunden erreicht die Spannung beim Laden ca. 63% der Endspannung</p>
<p>- Nach 5·τ Sekunden ist der Kondensator fast vollständig geladen</p>
""")

theorie_box_tau = widgets.VBox([theorie_html_tau])
theorie_box_tau.layout = widgets.Layout(
    width='45%',
    border='2px solid gray',
    padding='10px',
    margin='10px',
    min_height='300px',
    display='flex',
    align_items='flex-start'
)

# ----------------------------
# Gesamt-Layout nebeneinander
# ----------------------------
layout_tau = widgets.HBox([theorie_box_tau, rechner_box_tau],
                          layout=widgets.Layout(
                              justify_content='center',
                              align_items='flex-start',
                              width='100%'
                          ))

display(layout_tau)

# Initialer Aufruf für Anzeige
update_tau()
```
